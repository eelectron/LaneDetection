```python
import cv2
from matplotlib import pyplot as plt
```

```python
bgrImg = cv2.imread('./dataset/ld1.jpg')
print(bgrImg.shape)
b,g,r = cv2.split(bgrImg)
rgbImg = cv2.merge([r,g,b])

lt = rgbImg[330:370, 240:290]  #left top crop
lb = rgbImg[420:460, 70:120]  #left bottom crop

mt = rgbImg[330:370, 320:370]  #middle top crop
mb = rgbImg[390:430, 360:410]  #middle bottom crop

rt = rgbImg[330:370, 400:450]  #right top crop
rb = rgbImg[390:430, 590:640]  #right bottom crop

#plot it
plt.subplot(231), plt.imshow(lt)
plt.subplot(234), plt.imshow(lb)

plt.subplot(232), plt.imshow(mt)
plt.subplot(235), plt.imshow(mb)

plt.subplot(233), plt.imshow(rt)
plt.subplot(236), plt.imshow(rb)
```
